### Spain Electricity Shortfall Project

### Introduction

The government of Spain is considering an expansion of it's renewable energy resource infrastructure investments. As such, they require information on the trends and patterns of the countries renewable sources and fossil fuel energy generation. 
The purpose of the project is to model the shortfall between the energy generated by means of fossil fuels and various renewable sources - for the country of Spain. The daily shortfall, which will be referred to as the target variable, will be modelled as a function of various city-specific weather features such as `pressure`, `wind speed`, `humidity`, etc. As with all data science projects, the provided features are rarely adequate predictors of the target variable. As such, you are required to perform feature engineering to ensure that you will be able to accurately model Spain's three hourly shortfalls.

# Handling Outliers

* Outliers have a significant impact on the overall distribution of the data. 
* Handling outliers is a key step in data scaling with the aim of improving prediction accuracy. 
* The focus of this section is on the obvious outliers.

```python
# Checking for outliers.
df.kurtosis() 
```

* The fact that the maximum recorded pressure level in Barcelona is 1084 justifies the conclusion that there is an error with the outlier in Barcelona pressure. 
* Values greater than 1084 will be replaced with the median to avoid errors caused by too many outliers.
* The other unlikely value is the Valencia_wind_speed value of 32 because the highest recorded wind speed in spain is 20. 
* The recent significant changes in weather patterns and intensity can explain the the other outliers.

# Ridge Regularisation and Regression

```python
X_standardise = pd.DataFrame(X_scaled, columns=X.columns)
X_standardise.head()
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

# Instantiage Ridge model
ridge = Ridge()

# Train Model
ridge.fit(X_train, y_train)

```

```python
# Extract the model intercept value
b0 = float(ridge.intercept_)
```

```python
# Extract the model coefficient value
coeff = pd.DataFrame(ridge.coef_, X.columns, columns=['Coefficient'])
```

```python
print("Intercept:", float(b0))
```

```python
# Check out the coefficients
coeff
```

Application of Lasso Regularization method

```python
lasso=Lasso(alpha=0.8)

# Train Model
lasso.fit(X_train, y_train)

# Extract intercept
intercept_lasso = float(lasso.intercept_)

# Extract coeffiecient from model
coefficients = pd.DataFrame(lasso.coef_, X.columns, columns=['Coefficient'])

 # Extract intercept
print('Intercept: ', float(intercept_lasso))
```

```python
coefficients
```

* None of the features have been reduced to exactly zero.
* Increasing alpha from 0.01 to 1.0 to 1.5 reduced the tolerance and duality gap.
* The intercept is the expected load_shortfall when all the features are equal their means
* The use of Lasso regularization has made it clear that all of the features in the dataset can influence the predictive quality of the model.

# Linear Regression

```python
lm = LinearRegression()

# train model
lm.fit(X_train, y_train)
```

# Check Training Accuracy

```python
train_lm = lm.predict(X_train)
train_ridge = ridge.predict(X_train)
train_lasso = lasso.predict(X_train)

print('Training RMSE')
print('Linear:', np.sqrt(MSE(y_train, train_lm)))
print('Ridge :', np.sqrt(MSE(y_train, train_ridge)))
print('Lasso: ', np.sqrt(MSE(y_train, train_lasso)))
```

# Test MSE( linear, ridge and lasso)

```python
test_lm = lm.predict(X_test)
test_ridge = ridge.predict(X_test)
test_lasso = lasso.predict(X_test)

print('Testing MSE')
print('Linear:', np.sqrt(MSE(y_test, test_lm)))
print('Ridge :', np.sqrt(MSE(y_test, test_ridge)))
print('Lasso: ', np.sqrt(MSE(y_test, test_lasso)))
```

```python
def hyperParameterTuning(X_train, y_train):
    param_tuning = {
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5, 7, 10],
        'min_child_weight': [1, 3, 5],
        'subsample': [0.5, 0.7],
        'colsample_bytree': [0.5, 0.7],
        'n_estimators' : [100, 200, 500],
        'objective': ['reg:squarederror']
    }

    xgb_model = XGBRegressor()

    gsearch = GridSearchCV(estimator = xgb_model,
                           param_grid = param_tuning,                        
                           #scoring = 'neg_mean_absolute_error', #MAE
                           #scoring = 'neg_mean_squared_error',  #MSE
                           cv = 5,
                           n_jobs = -1,
                           verbose = 1)

    gsearch.fit(X_train,y_train)

    return gsearch.best_params_


hyperParameterTuning(X_train, y_train)
```
